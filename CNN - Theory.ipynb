# Convolutional Neural Networks

## What are CNNs?
* Type of Neural Networks that are extremely effective at dealing with image data

## What is Computer Vision?
General term of using computer programs to process image data

To better understand this field we must understand:
* Image Filters and Kernels

## How do Image Filters work?
* Filters are **image kernels** i.e. a small matrix applied to an entire image
* Filters can transform images
* To apply certain filters, we would multiply the 3X3 matrix into every single pixel, then sum the matrix's total number and apply to a new image where the filter is applied. This process is called **Convolution**

## What is a Convolutional Layer?
* Layer created when we apply multiple image filters to input images
* Layer will figure the best weight values
* Not all neurons will be fully connected
* Neurons will only connect to a local neuron on the next layer

## What are Pooling Layers?
This type of layer will help us deal with a large amount of parameters:
* Color Images
* Images with more than 10 filters

This type of layer will end up removing a lot of information. 75% of data to be precise.

We can use Max Pooling: Which consists of taking the max value of a 2X2 matrix
